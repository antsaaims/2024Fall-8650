# Matrix demo
Let's see how long it takes to take a Frobenius norm

```python
import math
import time
```

```python
def make_matrix(N):
    matrix = [[0] * N for i in range(N)]
    return matrix
```

```python
def fill(matrix):
    N = len(matrix)
    for i in range (N):
        for j in range (N):
            matrix[i][j]=1.0*(i+j)
m=make_matrix(5)
fill(m)
print(m)            
```

```python
def frobenius(matrix):
    result = 0
    N = len(matrix)
    for i in range(N):
        for j in range(N):
            result += abs(matrix[i][j])**2
    return math.sqrt(result)
```

```python
# try 1000, 2000, 4000, 8000
matrix = make_matrix(4000)
fill(matrix)

t1=time.perf_counter()
print("Norm =", frobenius(matrix))
print("This took %.4fs" % (time.perf_counter()-t1))
```
